```python
def if_outlier(data,col):
    data[col].plot(kind='box', color='green')
```

```python
def remove_outlier(data,col):
    q1= data[col].quantile(0.25)
    q3= data[col].quantile(0.75)
    iqr= q3-q1
    u_lim= q3+1.5*iqr
    l_lim= q1-1.5*iqr
    data[col]= np.where(data[col]<l_lim, l_lim, data[col])
    data[col]= np.where(data[col]>u_lim, u_lim, data[col])
    if_outlier(data,col)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x2, y, test_size=0.2, random_state=10)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor()
```

```python
rf.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
print("MSE of training data=", mean_squared_error(y_train, rf.predict(x_train)))
print("MSE of testing data=", mean_squared_error(y_test, rf.predict(x_test)))
```

```python
for name, importance in zip(x2.columns, rf.feature_importances_):
    print("feature:", name, '=', importance)
```

```python
rf.score(x_test, y_test)
```
